```python
from pyspark.sql import (
    SparkSession,
    types,
    functions as F,
)

spark = (
    SparkSession
    .builder
    .appName('cleaning_orders_dataset_with_pyspark')
    .getOrCreate()
)
```

```python
orders_data = spark.read.parquet('data/orders_data_clean.parquet')
orders_data.toPandas().head()
```

```python
orders_data_glue = spark.read.parquet('data/awsglue_orders_data_processed.parquet')
orders_data_glue.toPandas().head()
```

```python
diff_df = orders_data.subtract(orders_data_glue)
diff_df.show()
```
